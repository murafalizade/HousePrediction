```python
import numpy as np
import pandas as pd
from sklearn.preprocessing  import LabelEncoder
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
```

```python
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import datetime
import time
chromrdriver = "C:/Users/HP/Downloads/chromedriver_win32 (1)/chromedriver.exe"
os.environ["webdriver.chrome.driver"] = chromrdriver
source = []
driver = webdriver.Chrome(chromrdriver)
for x in range(1,700):
    page = "https://bina.az/alqi-satqi?page"+"="+str(x)
    driver.get(page)

    ScrollNumber = 2
    for i in range(1,ScrollNumber):
        driver.execute_script("window.scrollTo(1,2000)")
        time.sleep(5)
    source.append(driver.page_source)


driver.close()
print(source)
```

```python
bina_soup = BeautifulSoup(source[0],'html.parser')
binas_prices = bina_soup.findAll('div',attrs={"class":"price"})
binas_loc = bina_soup.findAll('div',attrs={"class":"location"})
binas_names = bina_soup.findAll('ul',attrs={"class":"name"})
binas_dates = bina_soup.findAll('div',attrs={"class":"city_when"})
for i in range(1,len(source)):
    bina_soup = BeautifulSoup(source[i],'html.parser')
    binas_prices += bina_soup.findAll('div',attrs={"class":"price"})
    binas_loc += bina_soup.findAll('div',attrs={"class":"location"})
    binas_names += bina_soup.findAll('ul',attrs={"class":"name"})
    binas_dates += bina_soup.findAll('div',attrs={"class":"city_when"})
binas_dates
```

```python
names = []
locations = []
import numpy as np
prices = []
dates = []
for i in range(0,len(binas_names)):
    names.append(binas_names[i].text)
    locations.append(binas_loc[i].text)
    dates.append(binas_dates[i].text[binas_dates[i].text.find(" ")+1:len(binas_dates[i].text)])
    prices.append(int(binas_prices[i].text[0:binas_prices[i].text.find("A")].replace(" ",""))
    
area = []
floor = []
room = []
for x in names:
    i = x.find('o')
    j = x.find('ı')
    k= x.find('m')
    q = x.find("ə")
    rm = x[i-2:i]
    try:
        int(rm)
    except:
        rm = np.nan
    ar = x[j+1:k]
    try:
        float(ar)
    except:
        ar = ar[0:ar.find('s')]
    fl = x[k+2:q-1]
    if fl.find("s") != -1 or fl=="":
        floor.append(np.nan)
    else:
        floor.append(fl)
    area.append(ar)
    room.append(rm)
bins = pd.DataFrame({'Locations':locations,'Prices':prices,'Dates':dates,"Floor":floor,"Area":area,"Rooms":room})
#bins['Area'] = bins['Area'].astype(float)
#bins['Locations'].unique()
#bins['Rooms'].interpolate(method='linear')
bins
bins.to_csv('bnk1.csv')
```

```python
bins['Floor'] = bins['Floor'].fillna("1/1",axis=0)
bins['Rooms'] = bins['Rooms'].fillna(bins['Rooms'].mode()[0])
bins
```

```python
floors = []
current_floor = []
for x in np.array(bins['Floor']):
    current_floor.append(int(x.split('/')[0]))
    floors.append(int(x.split('/')[1]))
bins = bins.assign(Floors = floors,Current_floor = current_floor)
df = bins.drop(['Floor','Dates'],axis=1)
df
```

```python
df.to_csv('bnk2.csv')
```

```python
scaler = LabelEncoder()
df['LocationLabel'] = scaler.fit_transform(df['Locations'])
df['Rooms'] = scaler.fit_transform(df['Rooms'])
df
```

```python
from sklearn.model_selection import train_test_split
import xgboost as xg
from sklearn.metrics import accuracy_score
X = df.drop(['Locations','Prices'],axis=1)
y = df['Prices']
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=123,test_size=0.4)
xgb_r = xg.XGBRegressor(objective ='reg:linear',
                  n_estimators = 10, seed = 123)
  
# Fitting the model
xgb_r.fit(X_train, y_train)
  
# Predict the model
pred = xgb_r.predict(pd.DataFrame({'Area':[50.0],'Rooms':[3],'Floor':[2],'Current_floor':[5],"LocationLabel":[43]}))
print(pred)
```

```python
def predictBina(ar,rm,fl,cfl,ll):
    ll = np.array(df[df['Locations']==ll]['LocationLabel'])[0]
    pred = xgb_r.predict(pd.DataFrame({'Area':[ar],'Rooms':[rm],'Floor':[fl],'Current_floor':[cfl],"LocationLabel":[ll]}))
    return pred[0]
print(predictBina(100,4,15,10,'Yasamal q.'))
```
